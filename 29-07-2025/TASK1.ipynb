```python
import pandas as pd
import matplotlib.pyplot as plt


csv_data = """Name,Math,Science,English
Amit,78,85,74
Priya,92,89,96
Kiran,65,70,60
Anita,88,95,90
Ravi,55,60,58
"""


with open('students.csv', 'w') as f:
    f.write(csv_data)
```

```python
df = pd.read_csv('students.csv')
print("First 3 rows:")
print(df.head(3))
```

```python
df['Average'] = df[['Math','Science','English']].mean(axis=1)
print("\nStudent with highest average:")
print(df.loc[df['Average'].idxmax()])
```

```python
print("\nStudents with Science > 80:")
print(df[df['Science'] > 80])

print("\nStudents with Average < 70:")
print(df[df['Average'] < 70])
```

```python
print("\nTop 3 students by English score (descending):")
print(df.sort_values(by='English', ascending=False).head(3))
```

```python
plt.figure(figsize=(12,4))


plt.subplot(1,2,1)
plt.bar(df['Name'], df['Average'], color='skyblue')
plt.title('Students Average Scores')
plt.xlabel('Name')
plt.ylabel('Average Score')


plt.subplot(1,2,2)
df.set_index('Name')[['Math','Science','English']].plot(marker='o')
plt.title('Students Scores in Each Subject')
plt.ylabel('Scores')
plt.grid(True)
plt.show()
```

```python
subject_totals = df[['Math','Science','English']].sum()
highest_subject = subject_totals.idxmax()
print("\nSubject with highest total score:", highest_subject)


df.to_csv('students_summary.csv', index=False)
print("\nUpdated DataFrame saved as students_summary.csv")
```
